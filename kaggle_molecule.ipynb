```python
train = pd.read_csv('train.csv')
```

```python
train
```

```python
type_onehot = pd.get_dummies(train['type'])
```

```python
#model = xgb.sklearn.XGBRegressor()
model = lightgbm.sklearn.LGBMRegressor()
```

```python
model.fit(pd.concat([train[['atom_index_0','atom_index_1']],type_onehot],axis=1),train['scalar_coupling_constant'])
#model.fit(train[['atom_index_0','atom_index_1','type']],train['scalar_coupling_constant'])
```

```python
test = pd.read_csv('test.csv')
```

```python
test_type_onehot = pd.get_dummies(test['type'])
```

```python
result = model.predict(pd.concat([test[['atom_index_0','atom_index_1']],test_type_onehot],axis=1))
```

```python
submission = pd.concat([test[['id']],df(result,columns={"scalar_coupling_constant"})],axis=1)
```

```python
df(submission).to_csv('submission.csv',index=False)
```
